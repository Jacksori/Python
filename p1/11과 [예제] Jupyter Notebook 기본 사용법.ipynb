## 5. 코드 셀

### 1) 코드셀 라인 번호 추가

* 단축키 : L


```python
sample_list = []
sumsum = 0

for i in range(10) :
  sample_list.append(i)
  sumsum += i
  print(i)
print('loop done')
print(sumsum)
print(sample_list)
```
